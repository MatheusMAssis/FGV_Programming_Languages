```python
import math

#Definir se um número é primo ou não

def eprimo(x):
    eprimo = 1
    for i in range(2, int(sqrt(x))+1):
            if x % i == 0:
                eprimo = 0
                break
    return eprimo
```

```python
#Encontrar o n-ésimo primo
#Project Euler #7

def eneprimo(n):
    i = 2
    k = 3
    while i < n:
        k = k + 2
        if eprimo(k) == 1:
            i = i + 1
    return k

eneprimo(10001)
```

```python
#Realiza o produto dos números em uma lista

def prod(lista):
    prod = 1
    for i in range(len(lista)):
        prod = prod * lista[i]
    return prod
```

```python
#Project Euler #12
#1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

#DIVISORES
# 1: 1
# 3: 1,3
# 6: 1,2,3,6
#10: 1,2,5,10
#15: 1,3,5,15
#21: 1,3,7,21
#28: 1,2,4,7,14,28

def conta_fatores(n):
    c = 0
    pot = 0
    while n % 2 == 0:
        n = n/2
        c = c + 1
        pot = pot + 1
    i = 3
    while n > 1:
        while n % i == 0:
            n = n/i
            c = c + 1
            
        i = i + 2
    return c, pot


s = 0
k = 0
ndiv = 6
contagem = 1
while contagem < ndiv:
    k = k + 1
    s = s + k
    if s % 2 == 0:
        contagem = conta_fatores(s)

print(s)

conta_fatores()
```

```python
#Project Euler #25

def fibonacci_digit(n):
    f1 = 0
    f2 = 1
    f3 = f1 + f2
    for m in range(1,n-1):
        f1 = f2
        f2 = f3
        f3 = f1 + f2
    return f3

n = len(str(fibonacci_digit(12)))
k = 12
while n < 1000:
    k = k + 1
    n = len(str(fibonacci_digit(k)))
print(k)
```

```python
#Soma em lista

def soma(lista):
    soma = 0
    for i in range(len(lista)):
        soma = soma + lista[i]
    return soma
```

```python
#Resultado do fatorial de n

def fatorial(n):
    prod = 1
    while n >= 1:
        prod = prod * n
        n = n - 1
    return prod
```

```python
#Project Euler #15

def combinação(n,k):
    comb = (fatorial(n))/(fatorial(k)*fatorial(n-k))
    return comb

combinação(40,20)
```

```python
#Project Euler #14

def collatz(n):
    L = [n]
    while n > 1:
        if n % 2 == 0:
            n = n/2
        else:
            n = 3 * n + 1
        L.append(n)
    
    return len(L), L[0]

p = 0
n2 = 0
for i in range(1,1000000,2):
    k = collatz(i)[0]
    if k > n2:
        n2 = k
        p = collatz(i)[1]

print(p)
```
